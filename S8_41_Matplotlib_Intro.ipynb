# Functional method

```python
plt.plot(x,y,'r-')
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#00
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x,y)
axes1.set_title('LARGER PLOT')

axes2.plot(y,x)
axes2.set_title('SMALLER PLOT')
```

## Matplotlib - 2

```python
fig,axes = plt.subplots(nrows = 1, ncols = 2 )

axes[0].plot(x,y)
axes[1].plot(y,x)

for current_ax in axes:
    current_ax.plot(x,y)
#now axes is kinda list because of for loop
```

```python
fig,axes = plt.subplots(nrows = 1, ncols = 2 )

axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')

```

### Figure Size and DPI

```python
fig = plt.figure(figsize=(3,2),dpi=100)

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```
